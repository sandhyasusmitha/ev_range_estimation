```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('./volkswagen_e_golf.csv')
```

```python
def tryconvert(value, default, t):
    if t == 'numeric':
        try:
            return float(value)
        except (ValueError, TypeError):
            return default
    if t == 'str':
        try:
            return str(value)
        except (ValueError, TypeError):
            return default
```

```python
#trip distance is a float value
data['trip_distance(km)'] = data['trip_distance(km)'].map(lambda x : tryconvert(x, np.nan, 'numeric') )
data['avg_speed(km/h)'] = data['avg_speed(km/h)'].map(lambda x : tryconvert(x, np.nan, 'numeric') )
data['consumption(kWh/100km)'] = data['consumption(kWh/100km)'].map(lambda x : tryconvert(x, np.nan, 'numeric') )
data['quantity(kWh)'] = data['quantity(kWh)'].map(lambda x : tryconvert(x, np.nan, 'numeric') )
```

```python
feature = 'city'
data.groupby(feature)['trip_distance(km)'].mean().plot.bar()
plt.xlabel(feature)
plt.ylabel('Trip distance(Km)')
plt.title(feature)
plt.show()
```

```python
feature = 'motor_way'
data.groupby(feature)['trip_distance(km)'].mean().plot.bar()
plt.xlabel('Motor way')
plt.ylabel('Trip distance(Km)')
plt.title(feature)
plt.show()
```

```python
feature = 'country_roads'
data.groupby(feature)['trip_distance(km)'].mean().plot.bar()
plt.xlabel('Country roads')
plt.ylabel('Trip distance(km)')
plt.title(feature)
plt.show()
```

```python
feature = 'A/C'
data.groupby(feature)['trip_distance(km)'].mean().plot.bar()
plt.xlabel('A/C')
plt.ylabel('Trip distance(km)')
plt.title(feature)
plt.show()
```

```python
feature = 'park_heating'
data.groupby(feature)['trip_distance(km)'].mean().plot.bar()
plt.xlabel('Park Heating')
plt.ylabel('Trip distance(km)')
plt.title(feature)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.title("Average speeds vs city")
sns.violinplot(x = 'city', y = 'avg_speed(km/h)', data = data)

plt.subplot(2, 2, 2)
plt.title("Average speeds vs motor_way")
sns.violinplot(x = 'motor_way', y = 'avg_speed(km/h)', data = data)

plt.subplot(2, 2, 3)
plt.title("Average speeds vs country_roads")
sns.violinplot(x = 'country_roads', y = 'avg_speed(km/h)', data = data)

plt.subplot(2, 2, 4)
plt.title("Average speeds vs driving_style")
sns.violinplot(x = 'driving_style', y = 'avg_speed(km/h)', data = data)

plt.show()
```
